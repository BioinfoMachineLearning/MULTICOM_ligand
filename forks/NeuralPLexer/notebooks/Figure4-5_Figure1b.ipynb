```python
def violin_plot_df(to_plot, title, ylabel):
    fontsize = 10
    fig, axes = plt.subplots()
    sns.violinplot(data=to_plot, ax = axes, bw=.2)
    axes.set_title(title)

    axes.yaxis.grid(True)
    # axes.set_xlabel('Scenario')
    axes.set_ylabel(ylabel)

    plt.show()
```

```python
def box_plot_df(to_plot, title, ylabel):
    fontsize = 10
    fig, axes = plt.subplots(figsize=(6, 6))
    sns.boxplot(data=to_plot, ax = axes, width=0.5, flierprops={"marker": "."})
    axes.set_title(title)

    axes.yaxis.grid(True)
    # axes.set_xlabel('Scenario')
    axes.set_ylabel(ylabel)
    plt.show()
```

```python
def bar_plot_df(to_plot, title, ylabel):
    fig, axes = plt.subplots(figsize=(4, 3))
    sns.set(font_scale=1)
    custom_params = {"axes.spines.right": False, "axes.spines.top": False}
    sns.set_theme(style="ticks", rc=custom_params)
    sns.set_palette("coolwarm", n_colors=3)
    sns.barplot(data=to_plot, ax=axes, width=0.6)
    axes.set_title(title, fontsize=16)

    #axes.yaxis.grid(False)
    # axes.set_xlabel('Scenario')
    axes.set_ylabel(ylabel, fontsize=16)
    #sns.despine(bottom)
    plt.ylim(0.85, 0.96)
    plt.show()
```

```python
def get_df(ds_name, mode, sampler="langevin_simulated_annealing"):
    dfs = []
    df = pd.read_csv(f"../neuralplexermodels_downstream_datasets_predictions/predictions/{ds_name}/set0/trainfrac1/default/af2_template/{mode}/{sampler}_T1.0_NFE100/summary.csv")
    dfs.append(df)
    for i in range(1, 6):
        dfi = pd.read_csv(f"../neuralplexermodels_downstream_datasets_predictions/predictions/{ds_name}/msa512_set1to5/default/af2_template_msa512/{mode}/{sampler}_T1.0_NFE100/model{i}/summary.csv")
        dfs.append(dfi)
    return pd.concat(dfs)
```

```python
data_path = "../neuralplexermodels_downstream_datasets_predictions/predictions/pocketminer/msa512_set1to5/default/af2_template_msa512/inferred_prior/langevin_simulated_annealing_T1.0_NFE100"
control_path = "../neuralplexermodels_downstream_datasets_predictions/predictions/pocketminer/msa512_set1to5/default/af2_template_msa512/discard_ligand/langevin_simulated_annealing_T1.0_NFE100"
metadata_df = pd.read_csv("../datasets/refined_training_datasets/0221_pocketminer_3A_plddt70_paired_with_seqsim.csv").set_index("protein_sample_id")
pkmn_df = pd.read_csv("../datasets/pocketminer_test/processed_0829.csv")
q_factor_pred_res, q_factor_control_res, q_factor_af2_res = [], [], []
seqsim = []
CUTOFF = 20.0
min_motion_dist = 2.0
for i in range(1, 6):
    out_path = data_path+f"/model{i}"
    control_out_path = control_path+f"/model{i}"
    df = pd.read_csv(out_path+f"/summary.csv")
    df = df.sort_values('plddt', ascending=False).drop_duplicates("sample_id")
    holo_subset_filter = np.array(~df["ref_sdf"].isna())
    for _, row in tqdm.tqdm(df[holo_subset_filter].iterrows()):
        sample_id = row["sample_id"]
        run_idx = row['run_id']
        pkmn_id = sample_id[:-1].upper() + "_" + sample_id[-1]
        pkmn_holo_id = pkmn_df[(pkmn_df["apo_id"]==pkmn_id)|(pkmn_df["holo_id"]==pkmn_id)].iloc[0]["holo_id"]
        pkmn_apo_id = pkmn_df[(pkmn_df["apo_id"]==pkmn_id)|(pkmn_df["holo_id"]==pkmn_id)].iloc[0]["apo_id"]
        if pkmn_apo_id == pkmn_id:
            alt_id = ''.join(pkmn_holo_id.split('_'))
        elif pkmn_holo_id == pkmn_id:
            alt_id = ''.join(pkmn_apo_id.split('_'))
        else:
            raise ValueError

        rec_path = f"{training_structures_path}/{sample_id}/{sample_id}.pdb"
        try:
            template_path = f"{training_structures_path}/{alt_id}/{alt_id}.pdb"
            sample_template, _ = featurize_protein_and_ligands(None, template_path, None)
        except:
            try:
                alt_id = alt_id[:4].lower() + alt_id[4]
                template_path = f"{training_structures_path}/{alt_id}/{alt_id}.pdb"
                sample_template, _ = featurize_protein_and_ligands(None, template_path, None)
            except:
                try:
                    alt_id = alt_id[:4].upper() + alt_id[4]
                    template_path = f"{training_structures_path}/{alt_id}/{alt_id}.pdb"
                    sample_template, _ = featurize_protein_and_ligands(None, template_path, None)
                except:
                    continue
        seqsim.append(metadata_df.loc[sample_id, "max_training_seq_similarity"])
        sample_ref, _ = featurize_protein_and_ligands(None, rec_path, None)
        pred_pdb_path = os.path.join(out_path, sample_id, f"prot_{run_idx}.pdb")
        sample_pred, _ = featurize_protein_and_ligands(None, pred_pdb_path, None)
        q_factor = compute_nonconserved_q_factor(sample_ref, sample_template, sample_pred, contact_cutoff=CUTOFF, min_motion_dist=min_motion_dist)
        q_factor_pred_res.append(q_factor.item())
        
        control_pdb_path = os.path.join(control_out_path, sample_id, f"prot_{run_idx}.pdb")
        sample_control, _ = featurize_protein_and_ligands(None, control_pdb_path, None)
        q_factor_ctl = compute_nonconserved_q_factor(sample_ref, sample_template, sample_control, contact_cutoff=CUTOFF, min_motion_dist=min_motion_dist)
        q_factor_control_res.append(q_factor_ctl.item())

        template_ids = metadata_df.loc[sample_id]["AF2_tempate_sample_id"].split(",")
        best_template_idx = np.argmax(
            [int(v.split("_")[1]) for v in template_ids]
        )
        openfold_id = metadata_df.loc[sample_id, "openfold_id"]
        af2_query = f"{template_structures_path}/{openfold_id}_unrelaxed_*_model_{i}.pdb"
        af2_path = glob.glob(af2_query)[0]
        sample_af2, _ = featurize_protein_and_ligands(None, af2_path, None)
        q_factor_af2 = compute_nonconserved_q_factor(sample_ref, sample_template, sample_af2, contact_cutoff=CUTOFF, min_motion_dist=min_motion_dist)
        q_factor_af2_res.append(q_factor_af2.item())
```

```python
from Bio.PDB import PDBList, Select, PDBParser, PDBIO
import os

class ChainSelect(Select):
    def __init__(self, chain):
        self.chain = chain

    def accept_chain(self, chain):
        return chain.get_id() == self.chain

def download_pdb_chain(pdb_id, chain_id, output_filename):
    # Download PDB file
    pdbl = PDBList()
    pdb_filename = pdbl.retrieve_pdb_file(pdb_id, pdir='.', file_format='pdb')

    # Parse PDB file and extract specific chain
    parser = PDBParser()
    structure = parser.get_structure(pdb_id, pdb_filename)
    io = PDBIO()
    io.set_structure(structure)
    io.save(output_filename, ChainSelect(chain_id))

    return output_filename

downloaded_file = download_pdb_chain("3KZV", "A", "3KZVA.pdb")
print(f"Downloaded file: {downloaded_file}")
```
